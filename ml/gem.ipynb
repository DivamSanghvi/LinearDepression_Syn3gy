````python
import base64
import os
from google import genai
from google.genai import types


def generate():
    client = genai.Client(
        api_key=os.environ.get("GEMINI_API_KEY"),
    )

    model = "gemini-2.0-pro-exp-02-05"
    contents = [
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text="""generate a diagram for atomic structure showing energy levels of each shell
"""
                ),
            ],
        ),
        types.Content(
            role="model",
            parts=[
                types.Part.from_text(
                    text="""I am currently unable to generate images directly, as this feature is available only for early testers.

However, I can provide you with a text-based representation of an energy level diagram and guidance on how to create one, or even provide an ASCII art version. Which would you prefer?

For example, here's a simple text-based representation:

```
Energy Level Diagram:

        n = âˆž  ------- Ionization
            ^
            |  (Energy decreases as n decreases)
        n = 4  -------
            ^
            |
        n = 3  -------
            ^
            |
        n = 2  -------
            ^
            |
        n = 1  ------- Ground State
```

Alternatively, I could create a more detailed description you can use to draw your own diagram or give to someone to create it for you.  Let me know which route you'd like to take!
"""
                ),
            ],
        ),
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text="""generate a diagram for atomic structure showing energy levels of each shell
"""
                ),
            ],
        ),
        types.Content(
            role="model",
            parts=[
                types.Part.from_text(
                    text="""Okay, since I can't *draw* the image, I'll give you a very detailed description that you can use to create the diagram yourself (using drawing software, pen and paper, etc.). I'll include enough detail to make it clear, and I'll structure it so it's easy to follow.  I'll also describe different levels of detail you could include.

**Atomic Structure Energy Level Diagram Description**

**1. Basic Structure (Core Idea):**

*   **Vertical Axis (Energy):** Draw a vertical line representing the energy scale.  Label the bottom \"Lowest Energy\" (or \"Ground State\") and the top \"Highest Energy\" (or \"Ionization\").  Don't put"""
                ),
            ],
        ),
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text="""INSERT_INPUT_HERE"""
                ),
            ],
        ),
    ]
    generate_content_config = types.GenerateContentConfig(
        temperature=1,
        top_p=0.95,
        top_k=64,
        max_output_tokens=8192,
        response_mime_type="text/plain",
    )

    for chunk in client.models.generate_content_stream(
        model=model,
        contents=contents,
        config=generate_content_config,
    ):
        print(chunk.text, end="")


generate()
````
